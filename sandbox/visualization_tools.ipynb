```python
# plot histograms of freeboard differences between the classic and drift-aware approach

vmin = -15
vmax = 15
figsize=(6,10)
fontsize=14

shp_file = "/Volumes/data2/projects/chione/ArcticRegions_shp/NSIDC-0780_SeaIceRegions_NH_v1.0.shp"
seaice_reg = gpd.read_file(shp_file)

var = "snow_freeboard_corrected"
diff_var = "snow_freeboard"

source_list = sorted(glob.glob("/Volumes/data2/projects/chione/is2_netcdf/da/" + '*.nc'))
diff_list = sorted(glob.glob("/Volumes/data2/projects/chione/is2_netcdf/cv/" + '*.nc'))
out_dir = "/Volumes/data2/projects/chione/is2_plots/diff_histograms/"

#source_list = source_list[0:1]

for file in source_list:
    data  = xr.open_dataset(file,decode_times=False)
    time_string = re.search('nh-(.+?)-(.*).nc', os.path.basename(file)).group(1)
    if [i for i in diff_list if time_string in i]:
        data_diff = xr.open_dataset([i for i in diff_list if time_string in i][0],decode_times=False)
            
    data["diff"] = data[var]-data_diff[diff_var]
    scaling = 100

    df = data["diff"].to_dataframe().reset_index()
    gdf = gpd.GeoDataFrame(df["diff"],geometry=gpd.points_from_xy(df.longitude,df.latitude)).dropna(how='any').reset_index(drop=True)
    merged = gpd.sjoin(gdf, seaice_reg, how='left', op='within')
    
    x1 = merged["diff"]*scaling
    x2 = merged.query('Region == "E_Greenland"')["diff"]*scaling
    x3 = merged.query('Region == "Beaufort"')["diff"]*scaling
    
    matplotlib.rc_file_defaults()
    plt.style.use('dark_background')
    
    fig, axes = plt.subplots(nrows=3, sharex=True, figsize=figsize)
    for x, ax in zip([x1,x2,x3], axes):
        hist(x,time_string,vmin,vmax,ax=ax)
    
    plt.annotate(time_string[6:8]+'.'+time_string[4:6]+'.'+time_string[0:4],
                 xy=(0.03, 3.3),fontsize=fontsize, xycoords='axes fraction', color='white')
    
    plt.annotate('Mean = '+'%.1f' % np.mean(x1) +' cm', xy=(0.6, 3.3),
                 fontsize=fontsize, xycoords='axes fraction', color='white')
    plt.annotate('Mean = '+'%.1f' % np.mean(x2) +' cm', xy=(0.6, 2.1),
                 fontsize=fontsize, xycoords='axes fraction', color='white')
    plt.annotate('Mean = '+'%.1f' % np.mean(x3) +' cm', xy=(0.6, 0.9),
                 fontsize=fontsize, xycoords='axes fraction', color='white')
    
    plt.annotate('Pan-Arctic',xy=(0.03, 3.1), fontsize=fontsize, xycoords='axes fraction', color='white')
    plt.annotate('E. Greenland',xy=(0.03, 1.9), fontsize=fontsize, xycoords='axes fraction', color='white')
    plt.annotate('Beaufort Sea',xy=(0.03, 0.7),fontsize=fontsize, xycoords='axes fraction', color='white')
    
    
    outfile = out_dir + re.split('.nc', os.path.basename(file))[0] + '_' + var + '_diff.png'
    plt.close(fig)
    fig.savefig(outfile, bbox_inches='tight',pad_inches=0, dpi=120)

matplotlib.rc_file_defaults()

```
